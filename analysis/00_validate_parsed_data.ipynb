```python
from pathlib import Path

import pandas as pd

ROOT = Path("..").resolve()
RESULTS = ROOT / "results" / "ground_state"
```

```python
validation = pd.read_csv(
    RESULTS / "validation" / "wavefunction_summary.csv"
)

orbitals = pd.read_csv(
    RESULTS / "orbitals" / "frontier_orbital_energies.csv"
)

charges = pd.read_csv(
    RESULTS / "charges" / "hirshfeld_atomic_charges.csv"
)

cubes = pd.read_csv(
    RESULTS / "cubes" / "cube_metadata.csv"
)

esp_metrics = pd.read_csv(
    RESULTS / "esp" / "esp_surface_metrics.csv"
)

esp_extrema = pd.read_csv(
    RESULTS / "esp" / "esp_extrema_from_text.csv"
)

mayer = pd.read_csv(
    RESULTS / "bond_orders" / "mayer_bond_orders.csv"
)

qtaim_cps = pd.read_csv(
    RESULTS / "qtaim" / "critical_points.csv"
)

qtaim_props = pd.read_csv(
    RESULTS / "qtaim" / "critical_point_properties.csv"
)

dos_curves = pd.read_csv(
    RESULTS / "dos" / "dos_curves.csv"
)

dos_lines = pd.read_csv(
    RESULTS / "dos" / "dos_lines.csv"
)
```
